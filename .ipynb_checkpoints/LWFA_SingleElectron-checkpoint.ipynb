```python
# 4/2/17
# Motion of a single electron with incident E&M plane wave

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
class LWFA(object):
    
    def __init__(self,omega=2):
        
        D = {}
        D['e'] = 1
        D['m'] = 1
        D['omega'] = omega
        D['omega_0'] = 1
        D['k'] = np.pi
        
        self.D = D
        
    def E0(self,z):
        
        D = self.D
        k = D['k']
        
        E0 = np.sin(k*z)
        
        return E0

    def x(self,t,z):
        
        D = self.D
        
        w = D['omega']
        w0 = D['omega_0']
        
        x = D['e']*self.E0(z)/(D['m']*(w**2-w0**2))*np.cos(w*t-np.pi/2)
        
        return x
```

```python
lwfa = LWFA()
    
# Position of one electron
x = lambda t: lwfa.x(t,1)
    
t = np.linspace(0,1,100)
position = [x(it) for it in t]
    
plt.plot(t,position,label="Position of a single electron")
plt.legend()
plt.xlabel('t [sec]')
plt.ylabel('Position on x-axis [arbitrary units]')
```

```python
lwfa = LWFA(omega=np.pi/2)

# Position of one electron
x_func = lambda t,z: lwfa.x(t,z)

# Position of several electrons along z-axis

z_array = np.linspace(0,1,50)
t_list = [0.0,0.33,0.67,1.0]
x_data = []

for it in t_list:
    
    x_pos_list = [x_func(it,iz) for iz in z_array]
    x_neg_list = [-ix for ix in x_pos_list]
    x_data.append(x_pos_list)
    x_data.append(x_neg_list)
```

```python
plt.hold(True)
plt.scatter(z_array,x_data[0],color='red',marker='.',label='t=0 sec')
plt.scatter(z_array,x_data[1],color='red',marker='.')
plt.scatter(z_array,x_data[2],color='blue',marker='.',label='t=0.33 sec')
plt.scatter(z_array,x_data[3],color='blue',marker='.')
plt.scatter(z_array,x_data[4],color='green',marker='.',label='t=0.67 sec')
plt.scatter(z_array,x_data[5],color='green',marker='.')
plt.scatter(z_array,x_data[6],color='black',marker='.',label='t=1.0 sec')
plt.scatter(z_array,x_data[7],color='black',marker='.')
#plt.ylim([-0.08,0.08])
plt.xlim([-0.2,1.4])
plt.ylabel('x [arbitrary units]')
plt.xlabel('z [arbitrary units]')
plt.legend()
plt.title('Ellipse of outer electrons as bubble evolves in time')
plt.show()
```
